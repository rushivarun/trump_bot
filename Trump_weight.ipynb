```
import numpy as np
```

```
# load ascii text and covert to lowercase
filename = "trump_input.txt"
raw_text = open(filename).read()
raw_text = raw_text.lower()
```

```
raw_text
```

```
chars = sorted(list(set(raw_text)))
print(len(chars))
print(chars)
```

```
# label encoding
char_num = dict((c, i) for i, c in enumerate(chars))
print(char_num)
```

```
char_num['a']
```

```
num_char = dict((i,c) for i, c in enumerate(chars))
num_char[32]
```

```
len_chars = len(raw_text)
len_vocabulary = len(chars)
print(len_chars)

print(len_vocabulary)
```

```
len_seq = 100
current_seq = []
next_char = []
for i in range(0, len_chars-len_seq, 1):
  seq_in = raw_text[i:i+len_seq]
  seq_out = raw_text[i+len_seq]
  
  current_seq.append([char_num[char] for char in seq_in])
  next_char.append(char_num[seq_out])
n_patterns = len(current_seq)
print(len(next_char))
```

```
print(len(current_seq[0]))
print(current_seq[1])
print(next_char[0])
```

```
import numpy as np 
from keras.utils import np_utils

X = np.reshape(current_seq, (n_patterns, len_seq, 1))
X = X/float(len_vocabulary)
Y = np_utils.to_categorical(next_char)
```

```
print(X.shape)
print(Y.shape)
```

```
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Dropout
```

```
model = Sequential()
model.add(LSTM(256, input_shape = (X.shape[1], X.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256))
model.add(Dense(Y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])
print(model.summary())
```

```
model.fit(X, Y, epochs =1)
```
